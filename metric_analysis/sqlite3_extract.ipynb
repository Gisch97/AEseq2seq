```python
import pandas as pd
import sqlite3

# Conectar a la base de datos SQLite
DATABASE_PATH = "../mlruns.db"
```

```python
experiment = "'seq2seq_plain'"
with sqlite3.connect(DATABASE_PATH) as conn:
    # Consulta SQL para pivotear la tabla directamente
    query = f"""
    SELECT * 
    FROM view_params AS p 
    JOIN view_metrics AS m ON p.run_uuid = m.run_uuid
    WHERE p.experiment_name = {experiment}
    """

    # Ejecutar la consulta y cargar los resultados en un DataFrame
    df_fetch = pd.read_sql_query(query, conn)


# Visualizar la tabla resultante
display(df_fetch)
```
